```python
import matplotlib.pyplot as plt
# %matplotlib notebook
%matplotlib inline
import pickle
import numpy as np
```

```python
def load_and_compute(mechanism,n_agent,reward_pool,review_history,window):
    filename="mechanism={}_n_agent={}_reward_pool={}_review_history={}_window={}.pkl".format(mechanism, n_agent, reward_pool,review_history,window )

    with open("./data/"+filename,"rb") as f:
        dict_=pickle.load(f)

    all_returns = np.array(dict_['all_returns'])
    all_costs = np.array(dict_['all_costs'])
    all_rewards = np.array(dict_['all_rewards'])
    all_actions = np.array(dict_['all_actions'])
    all_highests = np.array(dict_['all_highests'])
    all_likes = np.array(dict_['all_likes'])

    total_endeavor_ratio = np.sum(all_highests,axis=2) / (9*n_agent)
    average_total_endeavor_ratio = np.mean(total_endeavor_ratio,axis=0)
    n_episode = len(average_total_endeavor_ratio)

    return n_episode,average_total_endeavor_ratio

```

## exponential

```python
mechanism= "exponential"
n_agent= 100
reward_pool= 500
review_history= 2
window = 5
```

```python
n_episode_exp, average_total_endeavor_ratio_exp = load_and_compute(mechanism,n_agent,reward_pool,review_history,window)
```

```python
n_episode_exp
```

```python
average_total_endeavor_ratio_exp.shape
```

```python
plt.plot(np.arange(n_episode_exp),average_total_endeavor_ratio_exp)
```

## proportional

```python
mechanism= "proportional"
n_agent= 100
reward_pool= 500
review_history= 2
window = 5
```

```python
n_episode_pro, average_total_endeavor_ratio_pro = load_and_compute(mechanism,n_agent,reward_pool,review_history,window)
```

```python
mechanism= "uniform"
n_agent= 100
reward_pool= 500
review_history= 2
window = 5
```

```python
n_episode_uni, average_total_endeavor_ratio_uni = load_and_compute(mechanism,n_agent,reward_pool,review_history,window)
```
